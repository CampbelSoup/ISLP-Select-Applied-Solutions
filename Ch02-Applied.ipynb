**Problem 8**


```python
college = pd.read_csv("Data/College.csv")
college
```

```python
college2 = college.iloc[:, 0]
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
fig, ax = subplots(figsize = (8,8))
college.boxplot("Outstate", by = "Private", ax = ax)
```

```python
college['Elite'] = pd.cut(college['Top10perc'], bins = [0,0.5,1], labels = ['No', 'Yes'])
pd.value_counts(college["Elite"])
```

```python
fig, ax = subplots(2, 2, figsize = (8,8))
college["Grad.Rate"].plot.hist(ax=ax[0,0])
college['Accept'].plot.hist(ax = ax[0,1])
college["Personal"].plot.hist(ax=ax[1,0])
college["Top10perc"].plot.hist(ax=ax[1,1])  
```
